## Linear Search

```c++
#include <iostream>
#include <vector>

using namespace std;

// FUNCTION PROTOTYPES  (note lack of optional parameter names; if used, they don't have to match the DEFINITION names)

void fillVector (vector<int>&);  //  pass by ref to avoid mem copy
void printVector (const vector<int>&);  // note const is more mem efficient
bool searchVector (int, vector<int>&); //boolean-function for checking whether the number already exists or not
```

### Function Definitions

```c++
void fillVector(vector<int>& newMyVector) 
{
	int input; 
	cout << "Please type in the numbers and stop with (-1): ";
	cin >> input;


	while (input != -1) 
	{
		newMyVector.push_back(input);  // vector class public member that inserts objects at the end of the dynamic array
		cin >> input;
	}

}
```

```c++
void printVector(const vector<int>& newMyVector) 
{
	cout << "Array size: " << newMyVector.size() << endl;
	cout << "Array element 1: " << newMyVector[1] << endl;
	cout << "The existing numbers are: ";
	for (int i = 0; i < newMyVector.size(); i++) 
	{
		cout << newMyVector[i] << " ";
	}
	cout << endl;
}
```

```c++
bool searchVector (int enteredNumber, vector<int>& newMyVector) 
{	
	for (int i = 0; i < newMyVector.size(); i++) 
	{
		if (enteredNumber == newMyVector[i]) 
		{
			return true;
		}
	}
		return false;
}
```

### Execution (this would be main {} in an application code)

```c++
	vector<int> myVector;
	fillVector(myVector);
	printVector(myVector);

	int enteredNumber = 0;
	cout << "Please type in an int to compare with the numbers in the list: ";
	cin >> enteredNumber;

	if (searchVector(enteredNumber, myVector) == true ) 
	{
		cout << "The number already exists!" << endl;
	}

	else 
		cout << "The number was not typed in before!" << endl;
```
